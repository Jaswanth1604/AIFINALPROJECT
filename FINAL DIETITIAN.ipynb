```python
from tkinter import *
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd
from sklearn.tree import DecisionTreeClassifier


def getMeal():
    global prediction
    global e
    df = pd.read_csv('Diet.csv', encoding='cp1252')

    # print(df.head())

    X_train = df.loc[:, 'Gender':'Exercise']
    Y_train = df.loc[:, 'Diet']

    # The actual decision tree classifier
    tree = DecisionTreeClassifier(max_leaf_nodes=5, random_state=0)

    # Train the model
    tree.fit(X_train, Y_train)

    if Gender == 'Male':
        G = 1
    else:
        G = 2
    if Exercise == 'Sedentary':
        e = 1
    elif Exercise == 'Lightly Active':
        e = 2
    elif Exercise == 'Moderate':
        e = 3
    elif Exercise == 'Very Active':
        e = 4
    elif Exercise == 'SuperActive':
        e = 5
    prediction = tree.predict([[G, Age, e]])

    # print('Printing the prediction: ')

    # print(prediction)


def reset():
    txt.delete(0.0, END)
    e1.delete(0, END)
    rd1.deselect()
    rd2.deselect()
    e2.delete(0, END)
    e3.delete(0, END)
    e4.delete(0, END)
    e5.delete(0, END)
    spin.delete(0, END)
    c.set('Select')


def getDiet():
    global cal
    global Name
    global Gender
    global Exercise
    global Age
    global Height
    global Weight
    global cal

    txt.delete(0.0, END)
    Name = e1.get()
    if var_chk.get() == 1:
        Gender = "Male"
    else:
        Gender = "Female"
    Age = spin.get()
    Height = e2.get()
    Weight = e3.get()
    Address = e4.get()
    Mailid = e5.get()
    Exercise = c.get()

    if Gender == 'Male':
        cal = float()
        cal = 88.362 + (13.397 * float(Weight)) + (4.799 * float(Height)) - (5.677 * float(Age))
        # print(cal)
    elif Gender == 'Female':
        cal = float()
        cal = 447.593 + (9.247 * float(Weight)) + (3.098 * float(Height)) - (4.330 * float(Age))

    if Exercise == 'Sedentary':
        cal = cal * 1.2

    elif Exercise == 'Lightly Active':
        cal = cal * 1.375

    elif Exercise == 'Moderate':
        cal = cal * 1.55

    elif Exercise == 'Very Active':
        cal = cal * 1.725

    elif Exercise == 'SuperActive':
        cal = cal * 1.9
    getMeal()
    txt.insert(0.0, 'Meal Plan ' + prediction + '\n')
    txt.insert(0.0, '\n')
    txt.insert(0.0, "Calories: " + str(cal) + '\n')
    txt.insert(0.0, "MailId: " + Mailid + '\n')
    txt.insert(0.0, "Address: " + Address + '\n')
    txt.insert(0.0, "Exercise: " + Exercise + '\n')
    txt.insert(0.0, 'Weight: ' + Weight + 'Kgs' + '\n')
    txt.insert(0.0, 'Height: ' + Height + 'cms' + '\n')
    txt.insert(0.0, 'Age: ' + Age + 'Years' + '\n')
    txt.insert(0.0, 'Gender: ' + Gender + '\n')
    txt.insert(0.0, 'Name: ' + Name + '\n')


def mainWindow():
    global screen1
    global L1
    global f1
    global f2
    global l1
    global l2
    global l3
    global l4
    global l5
    global l6
    global l7
    global e1
    global e2
    global e3
    global e4
    global e5
    global rd1
    global rd2
    global txt
    global spin
    global droplist
    global Address
    screen1 = Tk()
    screen1.title("Dietitian")
    screen1.geometry("800x420")
    screen1.configure(bg='#3c513f')
    screen1.resizable(FALSE, FALSE)

    L1 = Label(screen1, height=1, width=60, font=("Berlin Sans FB", 20), text="DIETITIAN APPLICATION",compound = LEFT, fg="White", bg="#4aae56")
    L1.grid(row=1, column=1)
    # user Details
    f1 = Frame(screen1,bg='#cbdf0d')

    l1 = Label(f1, text='Name:', font=('Calibri', 14),bg='#cbdf0d')
    l2 = Label(f1, text='Gender:', font=('Calibri', 14),bg='#cbdf0d')
    l3 = Label(f1, text='Age:', font=('Calibri', 14),bg='#cbdf0d')
    l4 = Label(f1, text='Height', font=('Calibri', 14),bg='#cbdf0d')
    l5 = Label(f1, text='Weight', font=('Calibri', 14),bg='#cbdf0d')
    l6 = Label(f1, text='Exercise', font=('Calibri', 14),bg='#cbdf0d')
    l7 = Label(f1, text='Address', font=('Calibri', 14),bg='#cbdf0d')
    l8 = Label(f1, text="Mail id",font=('Calibri', 14),bg='#cbdf0d')


    e1 = Entry(f1, relief=SUNKEN)
    e2 = Entry(f1, relief=SUNKEN)
    e3 = Entry(f1, relief=SUNKEN)
    e4 = Entry(f1, relief=SUNKEN)
    e5 = Entry(f1, relief=SUNKEN)

    spin = Spinbox(f1, from_=18, to=60)

    global var_chk
    var_chk = IntVar()
    rd1 = Radiobutton(f1, text='Male', variable=var_chk, value=1,bg='#cbdf0d')
    rd2 = Radiobutton(f1, text='Female', variable=var_chk, value=2,bg='#cbdf0d')


    list1 = ['Sedentary', 'Lightly Active', 'Moderate', 'Very Active', 'SuperActive']
    global c
    c = StringVar()
    droplist = OptionMenu(f1, c, *list1)
    droplist.config(width=15)
    c.set('Select')

    b1 = Button(f1, text="Submit", bg='#15bfba', fg='black', font=('Calibri', 12), command=getDiet)

    l1.grid(row=0)
    l2.grid(row=1)
    l3.grid(row=2)
    l4.grid(row=3)
    l5.grid(row=4)
    l6.grid(row=5)
    l7.grid(row=6)
    l8.grid(row=7)
    

    e1.grid(row=0, column=1, padx='2')
    e2.grid(row=3, column=1, padx='2')
    e3.grid(row=4, column=1, padx='2')
    e4.grid(row=6, column=1, padx='2')
    e5.grid(row=7, column=1, padx='2')

    spin.grid(row=2, column=1)

    rd1.grid(row=1, column=1, sticky=W, padx='2')
    rd2.grid(row=1, column=1, sticky=E)

    droplist.grid(row=5, column=1)
    b1.grid(row=8, column=1)

    f1.grid(row=2, column=1, sticky=W)

    f2 = Frame(screen1, relief=SUNKEN,bg='#000000')

    photo2 = PhotoImage(file=r"diet1.png")
    p2 = photo2.subsample(8)
    l7 = Label(f2,width=400, text='Your Diet Plan', image = p2, font=('Calibri', 16),bg='#cbdf0d',fg='Black',compound = LEFT)
    l7.grid(row=0)

    txt = Text(f2, width=50, height=15, wrap=WORD, relief=SUNKEN)

    txt.insert(0.0, 'SuperActive:twice/day')
    txt.insert(0.0, 'VeryActive:6-7day/week' + '\n')
    txt.insert(0.0, 'Moderate:3-5day/week' + '\n')
    txt.insert(0.0, 'LightlyActive:1-3day/week' + '\n')
    txt.insert(0.0, 'Sedentary:No Exercise' + '\n')
    txt.insert(0.0, "" + '\n')
    txt.insert(0.0, 'Please Enter Height In Cms:' + '\n')
    txt.insert(0.0, 'Please Enter Weight In Kgs:' + '\n')
    txt.insert(0.0, 'Please Enter Age Between 18 To 60 Years only:' + '\n')
    txt.grid(row=1)

    b2 = Button(f2, text="Reset", bg='Red', fg='black', font=('Calibri', 12), command=reset)
    b2.grid(row=2)

    f2.grid(row=2,column=1)
    screen1.mainloop()


mainWindow()
```
